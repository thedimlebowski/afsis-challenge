```python
#!/usr/bin/python
import numpy as np
from NeuralNetworkOptimisation import NeuralNetworkOptimisation as NNO
import matplotlib.pyplot as plt
```

```python
n_points = 20
noise = 0.2
gap = 2*np.pi
train_in = np.array([list(np.linspace(0,5*np.pi-gap/2,n_points/2))+list(np.linspace(5*np.pi+gap/2,10*np.pi,n_points/2))]).T
train_out = np.array(np.sin(train_in) + noise*np.random.randn(len(train_in),1))
```

```python
n_sample = np.shape(train_in)[0]
n_input = np.shape(train_in)[1]
n_output = np.shape(train_out)[1]
net_arch = (n_input, 10  , n_output)
print net_arch
nno = NNO(net_arch, verbose = True, biases = True)
```

```python
nno.train_cv(train_in,train_out, ncv = 10, ntry = 10, ratio = 0.75, nproc = 4, best_test = False)
nno.train_best(train_in,train_out, ntry = 10, nproc = 4)
```

```python
outer_domain = 2*np.pi

plt.figure(figsize = (15,6))
test_in = np.array([np.linspace(0-outer_domain,10*np.pi+outer_domain,1000)]).T
test_out = np.array(np.sin(test_in))

colors = 'rgbcmyk'*10

test_net_out_best = nno.sim(test_in, best = True)
plt.plot(test_in,test_net_out_best, lw = 10, color= 'm',alpha = 0.8, label = 'BEST')

test_net_out_cv = nno.sim(test_in, cv = True)
plt.plot(test_in,test_net_out_cv, lw = 10, color= 'r',alpha = 0.8, label = 'CV average')

for i in range(nno.ncv)[::-1]:
    if i == 0:
        legend = 'CVs'
    else:
        legend = '_nolegend_'
    plt.plot(test_in,nno.sim(test_in, cv_i = i), lw = 2, color= colors[i],alpha = 0.3, label = legend, linestyle = '-')

plt.plot(test_in, np.sin(test_in),'--b', label = 'Real function')
plt.plot(train_in,train_out, lw = 0, color= 'b', marker = 'o', markersize = 20, label = 'Sampling')
plt.legend(loc = 0)

print 'cv', 0.5*(np.sum(np.abs(np.sin(test_in) - test_net_out_cv)**2)/n_points)
print 'best', 0.5*(np.sum(np.abs(np.sin(test_in) - test_net_out_best)**2)/n_points)

plt.show()
```
