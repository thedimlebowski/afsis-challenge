```python
train = pd.read_csv('training.csv').reset_index().rename(columns = {'index' : 'PID'})
nsamples = len(train)
wavelength_text = list(train.columns[2:-21])
wavelength = map(lambda x: float(x[1:]),wavelength_text)
train = train.rename(columns = dict(zip(wavelength_text,wavelength)))
params = list(train.columns[-21:-5])
params_to_predict = list(train.columns[-5:])
depth_map = {'Topsoil' : 1, 'Subsoil' : -1}
train['Depth'] = train['Depth'].apply(lambda x: depth_map[x])
spectra = train[wavelength+['PID']].set_index('PID').T.reset_index().rename(columns = {'index' : 'wavelength'})
```

```python
spectrum_ave = spectra.set_index('wavelength').T.mean().values
set_of_params = params_to_predict
plt.figure(figsize = (10,2*len(set_of_params)))
n = 0
corr = {}
all_corr=1
for param in set_of_params:
    n += 1
    corr[param] = []
    for i in range(len(wavelength)):
        corr[param]+=[np.abs(np.corrcoef(train[wavelength[i]],train[param])[0,1])]
    plt.subplot(len(set_of_params),1,n)
    if param in params_to_predict:
        color = 'r' 
    else:
        color = 'b'
    plt.plot(wavelength,corr[param], color = color)
    plt.plot(wavelength,spectrum_ave/max(spectrum_ave),color = [0.8]*3)
    plt.title(param)
    all_corr *= np.array(corr[param])
plt.subplots_adjust(hspace = 0.5)
plt.show()
plt.figure(figsize =(10,2))
plt.plot(wavelength,all_corr/max(all_corr), color = 'y')
plt.plot(wavelength,spectrum_ave/max(spectrum_ave),color = [0.8]*3)
```
